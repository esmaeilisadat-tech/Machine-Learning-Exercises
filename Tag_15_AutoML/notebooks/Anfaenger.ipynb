3. Create the best model, evaluate, and plot AUC/Confusion Matrix.

```python
# Uncomment to install if needed
# !pip install pycaret

import pandas as pd
from sklearn.datasets import load_breast_cancer
from pycaret.classification import setup, compare_models, create_model, evaluate_model, plot_model

# 1. Load Data and Setup
data = load_breast_cancer(as_frame=True)
df = data.frame

print("Setting up PyCaret environment. PyCaret automatically handles missing values, categorical encoding, train-test splitting, and feature scaling in the background.")
clf1 = setup(data=df, target='target', session_id=42, verbose=False)

# 2. Compare Models
print("Comparing all models...")
best_model = compare_models(sort='AUC')
print(f"\nBest Model: {best_model}")

# 3. Create & Evaluate
rf = create_model('rf') # Assuming Random Forest is top or near top
print("\nEvaluating Model (Interactive UI in Jupyter):")
# evaluate_model(rf) # Interactive

print("\nPlotting Confusion Matrix and AUC directly:")
plot_model(rf, plot='confusion_matrix')
plot_model(rf, plot='auc')
```
